### Exponential

```python
import math as m
def exp_pdf(x,l):
    if x>=0:
        return l*m.exp(-l*x)
    else:
        return 0
x=[i/10 for i in range(0,100)]
y1=[exp_pdf(i,l=0.5) for i in x]
y2=[exp_pdf(i,l=1) for i in x]
y3=[exp_pdf(i,l=1.5) for i in x]
y4=[exp_pdf(i,l=2) for i in x]
plt.plot(x,y1,label="l1")
plt.plot(x,y2,label="l2")
plt.plot(x,y3,label="l3")
plt.plot(x,y4,label="l4")
plt.legend()
plt.xlim(0,5)
plt.savefig("Exponential")
```

```python
import math as m
def exp_cdf(x,l):
    if x>=0:
        return 1-m.exp(-l*x)
    else:
        return 0
x=[i/10 for i in range(0,100)]
y1=[exp_cdf(i,l=0.5) for i in x]
y2=[exp_cdf(i,l=1) for i in x]
y3=[exp_cdf(i,l=1.5) for i in x]
y4=[exp_cdf(i,l=2) for i in x]
plt.plot(x,y1,label="l1")
plt.plot(x,y2,label="l2")
plt.plot(x,y3,label="l3")
plt.plot(x,y4,label="l4")
plt.legend()
```

### Normal Distribution

```python
import math
def nor_pdf(x, u, sd):
    sqrt_two_pi = math.sqrt(2*math.pi)
    return (math.exp(-(x-u)**2/2/ sd**2)/(sqrt_two_pi*sd))
x=[x/10.0 for x in range(-50,50)]
y1=[nor_pdf(i,u=0,sd=1) for i in x]
y2=[nor_pdf(i,u=0,sd=2) for i in x]
y3=[nor_pdf(i,u=0,sd=0.5) for i in x]
plt.plot(x,y1,label="u=0,sd=1")
plt.plot(x,y2,label="u=0,sd=2")
plt.plot(x,y3,label="u=0,sd=0.5")
plt.legend()       
```

## HYPOTHESIS TESTING

```python
def normal_cdf_below(x,u=0,sd=1):
    return(1+m.erf((x-u)/m.sqrt(2)/sd))/2
normal_cdf(8)
```

```python
def normal_cdf_above(x,u=0,sd=1):
    return (1-(1+m.erf((x-u)/m.sqrt(2)/sd))/2)
normal_cdf_above(8)
```

```python
def normal_cdf_between(x,u=0,sd=1):
    return (1-(1+m.erf((x-u)/m.sqrt(2)/sd))/2)
y1=normal_cdf_between(x=8)
y2=normal_cdf_between(x=9)
(y2-y1)
```

```python
def normal_cdf_outside(x,u=0,sd=1):
    return(1+m.erf((x-u)/m.sqrt(2)/sd))/2
y1=normal_cdf_outside(x=1)
y2=normal_cdf_outside(x=2)
1-(y2-y1)
```

```python
def inverse_cdf(p):
    high=10
    low=-10
    while(high-low)>0.0001:
        mid=(low+high)/2.0
        M=normal_cdf_below(mid)
        if M==p:
            return mid
        elif M>p:
            high=mid
        else:
            low=mid
    return high
inverse_cdf(0.5)   
```

```python
def normal_upper_bound(p,u=0,sd=1):
    return inverse_cdf(p)
normal_upper_bound(0.5)
```

```python
def normal_lower_bound(p,u=0,sd=1):
    return inverse_cdf(1-p)
normal_lower_bound(0.1586)
```

```python
def normal_two_sided_bounds(p,u=0,sd=1):
    tail_prob=(1-p)/2
    upper_bound=normal_lower_bound(tail_prob,u,sd)
    lower_bound=normal_upper_bound(tail_prob,u,sd)
    return lower_bound,upper_bound
normal_two_sided_bounds(0.13590)
```
