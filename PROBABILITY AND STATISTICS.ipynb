```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
def uniform_pdf(i):
    return 1 if i >=0 and i<=1 else 0
x=[i for i in range(-2,5)]
y=[uniform_pdf(z)for z in x]
print(y)
plt.plot(x,y,label='pdf')
```

```python
a=int(input("enter the value: "))
b=int(input("enter the value: "))
def uniform_pdf(i):
    return 1/(b-a) if i >=a and i<=b else 0
x=[i for i in range(a-5,b+5)]
y=[uniform_pdf(z)for z in x]
print(y)
plt.plot(x,y,label='pdf')
def uniform_cdf(x):
    if x<=a:
        return 0
    elif x<=b:
        return x
    else:
        return 1
y=[uniform_cdf(z)for z in x]
plt.plot(x,y,label='cdf')
plt.legend()
```

### Exponential

```python
import math as m
def exp_pdf(x,l):
    if x>=0:
        return l*m.exp(-l*x)
    else:
        return 0
x=[i/10 for i in range(0,100)]
y1=[exp_pdf(i,l=2) for i in x]
y2=[exp_pdf(i,l=1.5) for i in x]
y3=[exp_pdf(i,l=1) for i in x]
y4=[exp_pdf(i,l=0.5) for i in x]
plt.plot(x,y1,label="l1")
plt.plot(x,y2,label="l2")
plt.plot(x,y3,label="l3")
plt.plot(x,y4,label="l4")
plt.legend()
plt.xlim(0,5)
plt.savefig("Exponential")
```

```python
import math as m
def exp_cdf(x,l):
    if x>=0:
        return 1-m.exp(-l*x)
    else:
        return 0
x=[i/10 for i in range(0,100)]
y1=[exp_cdf(i,l=2) for i in x]
y2=[exp_cdf(i,l=1.5) for i in x]
y3=[exp_cdf(i,l=1) for i in x]
y4=[exp_cdf(i,l=0.5) for i in x]
plt.plot(x,y1,label="l1")
plt.plot(x,y2,label="l2")
plt.plot(x,y3,label="l3")
plt.plot(x,y4,label="l4")
plt.legend()
```

### Normal Distribution

```python
import math
def nor_pdf(x, u, sd):
    sqrt_two_pi = math.sqrt(2*math.pi)
    return (math.exp(-(x-u)**2/2/ sd**2)/(sqrt_two_pi*sd))
x=[x/10.0 for x in range(-50,50)]
y1=[nor_pdf(i,u=0,sd=1) for i in x]
y2=[nor_pdf(i,u=0,sd=2) for i in x]
y3=[nor_pdf(i,u=0,sd=0.5) for i in x]
plt.plot(x,y1,label="u=0,sd=1")
plt.plot(x,y2,label="u=0,sd=2")
plt.plot(x,y3,label="u=0,sd=0.5")
plt.legend()       
        
```
